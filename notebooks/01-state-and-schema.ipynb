# Step 1. State Schema

이 노트북은 `project.md`, `NOTE.md`, `ThingsToNote.md`를 기준으로
공통 State 구조와 운영 필드를 먼저 고정한다.


```python
from pathlib import Path
import sys
from dotenv import load_dotenv

CURRENT = Path.cwd().resolve()
PROJECT_ROOT = None
for candidate in [CURRENT, *CURRENT.parents]:
    if (candidate / "src" / "semiconductor_agent").exists():
        PROJECT_ROOT = candidate
        break
if PROJECT_ROOT is None:
    raise RuntimeError("project root with src/semiconductor_agent not found")

SRC_DIR = PROJECT_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.append(str(SRC_DIR))

for env_candidate in [PROJECT_ROOT / ".env", PROJECT_ROOT.parent / ".env"]:
    if env_candidate.exists():
        load_dotenv(env_candidate, override=False)
from semiconductor_agent.constants import COMPANIES, TARGET_TECHNOLOGIES, SEARCH_LIMITS
from semiconductor_agent.state import build_initial_state, snapshot_state_shape
```

## 1-1. Core State Fields

사용자 입력, 대상 기술, 경쟁사 범위를 확인한다.


```python
state = build_initial_state(
    "HBM4, PIM, CXL 기준 Samsung과 Micron의 최신 R&D 동향을 분석해줘."
)

print("technologies:", TARGET_TECHNOLOGIES)
print("companies:", COMPANIES)
print("search_limits:", SEARCH_LIMITS)
```

## 1-2. Evidence / Output / Ops Fields

state shape가 설계 문서와 맞는지 빠르게 확인한다.


```python
snapshot_state_shape(state)
```
